# Picture 1 - Loading and detecting faces.

```
import face_recognition as fr
```

```
image = fr.load_image_file('/content/drive/My Drive/data1.jpg')
```

```
from google.colab.patches import cv2_imshow
```

```
locs = fr.face_locations(image)
print("face detected ", len(locs))
```

```
print(locs)
```

```
import cv2 
```

```
for face in locs:
	   	cv2.rectangle(image,(face[3],face[0]),(face[1],face[2]),(0,225,0),2)
```

```
cv2_imshow(image)
```
